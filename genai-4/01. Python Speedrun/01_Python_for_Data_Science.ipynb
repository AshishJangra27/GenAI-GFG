### Operators

#### Arithmetic Operators

```python
print('1' + 3)
print('1' - 3)
print('1' * 3)
print('1' / 3)
print(1. // 3)
print(1. % 3)
print(1. ** 3)
```

#### Comparison

```python
print(3>5)
print(3<5)
print(3>=5)
print(3<=5)
print(3!=5)
print(3==5)
```

#### Logical

```python
age = 100

print((age>18) and (age<=100))
print((age>18) or (age<100))
print(not(True))
```

### Conditionals

```python
age = 25

if (age>=18 and age <= 100):
  print('Eligible')

elif (age>=0 and age < 18):
  print('Not Eligible')

else:
  print('Not a valid age')
```
